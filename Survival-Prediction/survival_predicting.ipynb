# USE LOGISTIC REGRESSION MODEL 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["class","deck","alive","who","adult_male","embarked"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"].fillna(df["age"].mean(),inplace=True)
df["age"]=df["age"].astype(int)
```

```python
df.dropna(subset=["embark_town"],inplace=True)
```

```python
df.head()
```

```python
df["embark_town"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["sex"]=le.fit_transform(df["sex"])
df["embark_town"]=le.fit_transform(df["embark_town"])
```

```python
df.head()
```

```python
df=df.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
H_scaler=StandardScaler()
df[["age","fare"]]=H_scaler.fit_transform(df[["age","fare"]])
```

```python
df
```

```python
y=df["survived"]
```

```python
df.drop(columns=["survived"],axis=1,inplace=True)
```

```python
X=df
X.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
